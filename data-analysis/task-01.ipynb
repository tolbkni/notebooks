### 任务一： 我们以"Age"为筛选条件，显示年龄在10岁以下的乘客信息。

### 任务二： 以"Age"为条件，将年龄在10岁以上和50岁以下的乘客信息显示出来，并将这个数据命名为midage

````python
midage = df[(df['Age'] > 10) & (df['Age'] < 50)]
midage.head()
````

### 任务三：将midage的数据中第100行的"Pclass"和"Sex"的数据显示出来
